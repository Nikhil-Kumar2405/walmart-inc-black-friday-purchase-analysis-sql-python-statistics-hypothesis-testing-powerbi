```python
# import necessaries libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sqlite3
warnings.filterwarnings('ignore')
```

```python
# import data from database
conn = sqlite3.connect('Purchase.db')

# fetching walmart black friday data
df = pd.read_sql_query('select * from walmart_data',conn)
df.head()
```

```python
# how many rows and columns
print(f'No. of rows {df.shape[0]}')
print(f'No. of columns {df.shape[1]}')
```

```python
# check data type
df.info()
```

```python
df.isnull().sum()/len(df)*100
```

```python
# summarize data
df.describe(include='all').T
```

```python
for col in df[['City_Category','Stay_In_Current_City_Years','Marital_Status','Product_Category']]:
    unique_values = df[col].unique()
    print('unique_values in: ',col,unique_values)
    
    
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.split('+').str[0].astype('int64')
```

```python
df['Gender'] = df['Gender'].replace({'M':'Male','F':'Female'})
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({0:'Married',1:'Unmarried'})
```

```python
df['Marital_Status'] = df['Marital_Status'].replace({'Unknown':'Unmarried'})
```

```python
# 'Old_Name' ko 'New_Name' se badalne ke liye
df = df.rename(columns={'Stay_In_Current_City_Years': 'Years'})
```

```python
df.info()
```

```python
age_mapping = {
    '0-17': 'Minor',
    '18-25': 'Young Adult',
    '26-35': 'Young Adult',
    '36-45': 'Adult',
    '46-50': 'Adult',
    '51-55': 'Senior',
    '55+': 'Senior'
}

# Naya column banayein
df['Age_Category'] = df['Age'].map(age_mapping)

# Ab iska count dekhein
print(df['Age_Category'].value_counts())
```

```python
df
```

```python
# converting columns name into lower cases with underscore
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df
```

```python
# saved manipulated data back into 'walmart_data'
walmart_datas = df.to_sql('walmart_data', conn, if_exists='replace', index=False)
print("Database updated with manipulated data!")
```
